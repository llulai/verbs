```python
import os
import json
from collections import Counter
```

```python
content = ''

for filename in os.listdir('news'):
    with open(f"news/{filename}", 'r') as file:
        content += file.read()
```

```python
count = Counter([w.lower() for w in content.split()])
```

```python
with open('conjugations.json', 'r') as file:
    conjugations = json.load(file)
```

```python
conjugations['abalar']
```

```python
def flatten(l):
    out = []
    for el in l:
        if isinstance(el, str):
            out.append(el)
        else:
            out.extend(flatten(el))
    return out


def get_childs(d):
    if isinstance(d, str):
        return d
    elif isinstance(d, dict):
        return [get_childs(v) for v in d.values()]
```

```python
flatten(get_childs(conjugations['abalar']))
```
